```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Name": ["Yersultan", "Kuanysh", "Arystan", "Danial", "Ruslan"],
    "Age": [19, 40, 15, 26, 5],
    "Score": [90, 75, 67, 34, 1]
}

df = pd.DataFrame(data)
```

```python
plt.figure()
plt.plot(df["Name"], df["Score"])
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Line Plot")
plt.show()
```

```python
plt.figure()
plt.bar(df["Name"], df["Score"])
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Bar Chart")
plt.show()
```
